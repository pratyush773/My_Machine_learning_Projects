```python
companies_dict={
'Amazon': 'AMZN',
'Apple': 'AAPL',
'Walgreen': 'WBA',
'Northrop Grumman': 'NOC',
'Boeing': 'BA',
'Lockheed Martin': 'LMT',
'McDonalds': 'MCD',
'Intel': 'INTC',
'Navistar': 'NAV',  
'IBM': 'IBM',
'Texas Instruments': 'TXN',
'MasterCard': 'MA',
'Microsoft': 'MSFT',
'General Electrics': 'GE',
'American Express': 'AXP',
'Pepsi': 'PEP',
'Coca Cola': 'KO',
'Johnson & Johnson': 'JNJ',
'Toyota': 'TM',
'Honda': 'HMC',
'Mitsubishi': 'MSBHY',
'Sony': 'SNE',
'Exxon': 'XOM',
'Chevron': 'CVX',
'Valero Energy': 'VLO',
'Ford': 'F',
'Bank of America': 'BAC'
}

companies = sorted(companies_dict.items(), key=lambda x: x[1])
print(companies)
```

```python
companies_dict.values()
```

```python
data_source='yahoo'
start_date='2015-01-01'
end_date='2017-12-31'
```

```python
panel_data=data.DataReader(list(companies_dict.values()), data_source, start_date, end_date) #.unstack(level=0)
```

```python
panel_data.head(2)
```

```python
print(panel_data.axes)
print(panel_data.shape)
panel_data.columns
```

```python
stock_open=panel_data['Open']
stock_close=panel_data['Close']
```

```python
print(stock_close.iloc[0])
```

```python
stock_close=np.array(stock_close).T
stock_open=np.array(stock_open).T
```

```python

print(stock_close.shape)
print(stock_open.shape)
```

```python
row,col = stock_close.shape
```

```python
movements = np.zeros([row, col])

for i in range(0, row):
    movements[i, : ] =  np.subtract(stock_close[i,:],stock_open[i, :])
```

```python
for i in range(0, len(companies)):
    print('Company {}: Change {}'.format(companies[i][0], sum(movements[i, : ])))
```

```python
movements.shape
```

```python
plt.clf
plt.figure(figsize=(18,16))
ax1=plt.subplot(221)
plt.plot(movements[0][:])
plt.title(companies[0])

ax2=plt.subplot(222, sharey=ax1)
plt.plot(movements[1][:])
plt.title(companies[1])

ax2=plt.subplot(223, sharey=ax1)
plt.plot(movements[2][:])
plt.title(companies[2])

ax2=plt.subplot(224, sharey=ax1)
plt.plot(movements[3][:])
plt.title(companies[3])

plt.show()
```

### why we need to normalise the data

```python
from sklearn.preprocessing import Normalizer

normalizer = Normalizer()
new = normalizer.fit_transform(movements)

print(new.max())
print(new.min())
print(new.mean())
```

```python
plt.clf
plt.figure(figsize=(18,16))
ax1=plt.subplot(221)
plt.plot(new[0][:])
plt.title(companies[0])

ax2=plt.subplot(222, sharey=ax1)
plt.plot(new[1][:])
plt.title(companies[1])

ax2=plt.subplot(223, sharey=ax1)
plt.plot(new[2][:])
plt.title(companies[2])

ax2=plt.subplot(224, sharey=ax1)
plt.plot(new[3][:])
plt.title(companies[3])

plt.show()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans
from sklearn.preprocessing import Normalizer

normaliser = Normalizer()
kmeans = KMeans(n_clusters = 10, max_iter = 1000)

pipeline = make_pipeline(normaliser, kmeans)
```

```python
pipeline.fit(movements)
```

```python
print(kmeans.inertia_)
```

```python
labels = pipeline.predict(movements)

df = pd.DataFrame({'lables' : labels, 'companies' : companies})

df.sort_values('lables')
```

```python
from sklearn.decomposition import PCA

reduced_data = PCA(n_components=2).fit_transform(new) # reduce the dimensioanlity using the PCA

kmeans = KMeans(init='k-means++', n_clusters = 10, max_iter = 1000) # fit the reduced data 
kmeans.fit(reduced_data)
labels = kmeans.predict(reduced_data)
```

```python
df = pd.DataFrame({'labels': labels, 'companies':companies})
print(df.sort_values('labels'))
reduced_data.shape
```

```python
h = 0.01
xmin, xmax = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
ymin, ymax = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(xmin,xmax, h), np.arange(ymin,ymax, h))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)

# Define Colormap
cmap = plt.cm.Paired

plt.figure(figsize=(10, 10))
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=cmap,
           aspect='auto', origin='lower')

plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=5)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='w', zorder=10)
plt.title('K-means clustering on Stock Market Movements (PCA-reduced data)\n'
          'Centroids are marked with white cross')
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.show()
```
